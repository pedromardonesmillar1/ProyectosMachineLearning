# Deep Dreams

El objetivo de este notebook es replicar el uso de una herramienta de [TensorFlow](https://www.tensorflow.org/tutorials/generative/deepdream) la cual está basada en el blog de [Alexander Mordvintsev](https://ai.googleblog.com/2015/06/inceptionism-going-deeper-into-neural.html) y la que consiste en la aplicación de un marco de desarrollo para la elaboración de una red neuronal con la cual se producen "sueños" por medio de la aplicación y del tratamiento de dichas redes neuronales. Este tratamiento permite "sobre implementar" y mejorar los puntos claves de la imagen que se desea visualizar la que entrega una imagen de "ensueño".

<img src = "https://b2h3x3f6.stackpathcdn.com/assets/landing/img/blend/horizontal/dd.jpg">

[Fuente: Deep Dream Generator](https://b2h3x3f6.stackpathcdn.com/)

Este procedimiento se lleva a cabo transfiriendo una imagen a través de la red neuronal y luego, calculando el [gradiente](https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/SGD) de dicha imagen, es que se generan las activaciones de las capas en particular que generaran esta imagen "sobre realista". Luego, la imagen se modifica para aumentar estas activaciones, mejorando los patrones vistos por la red y dando como resultado una imagen de ensueño. Este proceso se denominó "Inceptionism" (una referencia a [InceptionNet](https://towardsdatascience.com/a-simple-guide-to-the-versions-of-the-inception-network-7fc52b863202) y la película Inception).

#### Paso a paso para el desarrollo de esta red neuronal

### 1. Importar las librerias

Existe una gran cantidad de librerías que son utilizadas por Python para facilitar los cálculos y procedimientos que se necesitan llevar a cabo. En este caso, se utilizan los de mayor relevancia para el tratamiento de las imágenes.

```python
import tensorflow as tf
import numpy as np
import matplotlib as mpl
import PIL.Image
import IPython.display as display
from tensorflow import keras
```

### 2. Escoger la imagen que se desea tratar por la red neuronal

```python
# En el caso de querer descargar una imagen desde la web, esta imagen debe estar en formato 'imagen.jpg'.
url = 'https://fondosmil.com/fondo/13039.jpg'
```

```python
# El paso siguiente consiste en descargar la imagen y tratarla como una matriz numérica utilizando numpy.
def download(url, max_dim=None):
  name = url.split('/')[-1]
  image_path = tf.keras.utils.get_file(name, origin=url)
  img = PIL.Image.open(image_path)
  if max_dim:
    img.thumbnail((max_dim, max_dim))
  return np.array(img)

# Luego, el procedimiento se reduce a normalizar la imagen, es decir, desde los valores [0, 255] hasta [0, 1]. 
def deprocess(img):
  img = 255*(img + 1.0)/2.0
  return tf.cast(img, tf.uint8)

# Finalmente, la función 'show' permite mostrar en pantalla la imagen ya tratada.
def show(img):
  display.display(PIL.Image.fromarray(np.array(img)))
```

```python
# Con los pasos anteriores, la imagen se encuentra disponible para poder visualizarla, ya en formato numérico.
original_img = download(url, max_dim=500)
show(original_img)
```

### 3. Preparación del modelo para la extracción de sus características

Lo siguiente consiste en descargar y preparar un modelo de clasificación de imágenes el cual se encuentra previamente entrenado. Para este propósito es que se utilizará "InceptionV3", que es similar al modelo utilizado originalmente en DeepDream y que permite la generación de dichas imágenes de ensueño. Es necesario tener en cuenta que cualquier modelo previamente entrenado (no solamente con InceptionNet) funcionará, aunque será necesario ajustar los nombres de las capas en el modelo si es que se cambia o se modifica, ya sea utilizando solo partes del modelo, o el modelo completo.

```python
# Se carga el modelo Inception entrenado lo que permite obtener la información sobre el aprendizaje por transferencia.
base_model = tf.keras.applications.InceptionV3(include_top = False, weights = 'imagenet')
```

```python
# De esta manera, se puede visualizar el resumen de qué consiste la red neuronal en cuestión.
base_model.summary()
```

La idea en DeepDream es elegir una capa (o capas) y maximizar la "pérdida" de manera que la imagen "excite" cada vez más las capas. La complejidad de las características incorporadas depende de las capas elegidas, es decir, las capas inferiores producen trazos o patrones simples, mientras que las capas más profundas brindan características sofisticadas en las imágenes, o incluso en objetos completos.

La arquitectura InceptionV3 es bastante grande (para ver un gráfico de la arquitectura del modelo, lo recomendable es consultar [el repositorio de investigación de TensorFlow](https://github.com/tensorflow/models/blob/master/research/slim/nets/inception_v3.py)). Para DeepDream, las capas de interés son aquellas en las que se concatenan las circunvoluciones. Hay 11 de estas capas en InceptionV3, denominadas 'mixed0' hasta 'mixed10'. El uso de diferentes capas dará como resultado diferentes imágenes de ensueño. Las capas más profundas responden a funciones de nivel superior (como ojos y rostros), mientras que las capas anteriores responden a funciones más simples (como bordes, formas y texturas). Para el propósito de este trabajo, se utilizarán cinco capas de activaciones, que son 'mixed0', 'mixed3', 'mixed5', 'mixed7' y 'mixed9', no obstante, se puede trabajar con cualquier combinación de estas ya sea agregando más capas o quitando las que se encuentran numeradas. Sin embargo, es necesario tener en consideración que las capas más profundas (aquellas con un índice más alto) tardarán más en entrenarse ya que el cálculo del gradiente es más profundo.

```python
# Maximizar las activaciones de estas capas
names = ['mixed0', 'mixed3', 'mixed5', 'mixed7', 'mixed9']
layers = [base_model.get_layer(name).output for name in names]

# Crear las características de extracción del modelo
dream_model = tf.keras.Model(inputs=base_model.input, outputs=layers)
```

### Calcular pérdida (o función de costes)

La [pérdida](https://www.cienciadedatos.net/documentos/py35-redes-neuronales-python.html) es la suma de las activaciones en las capas elegidas. En resumen, es la diferencia que existe entre el valor observado (el real) y el valor proyectado (el visualizado por la red neuronal). A esta perdida (que es la diferencia entre ambas observaciones) entra en un proceso de tratamiento de la gradiente, donde se le conoce como el "aprendizaje" de la red, dado que es en esta etapa donde se espera que la diferencia observada (la suma de las perdidas en la función de costes) sea la mínima posible y que, en cada etapa o iteración, esta diferencia se acerque al valor de cero. 

La pérdida se normaliza en cada capa para que la contribución de las capas más grandes no supere a las capas más pequeñas. Normalmente, la pérdida es una cantidad que se desea minimizar a través del descenso de gradiente. En DeepDream, se maximizará esta pérdida a través del ascenso de gradiente.

```python
# Calcular las perdidas
def calc_loss(img, model):
  # Pasando la imagen por medio de las activaciones.
  # Convirtiendo la imagen a un lote de tamaño 1.
  img_batch = tf.expand_dims(img, axis=0)
  layer_activations = model(img_batch)
  if len(layer_activations) == 1:
    layer_activations = [layer_activations]

  losses = []
  for act in layer_activations:
    loss = tf.math.reduce_mean(act)
    losses.append(loss)

  return  tf.reduce_sum(losses)
```

### Ascenso de gradiente

Una vez que se haya calculado la pérdida de las capas elegidas, solo queda calcular los degradados con respecto a la imagen y agregarlos a la imagen original. Agregar los degradados a la imagen mejora los patrones vistos por la red. En cada paso de la imagen por la red neuronal, se habrá creado una imagen que excita cada vez más las activaciones de ciertas capas.

```python
# Gradiente descendente
class DeepDream(tf.Module):
  def __init__(self, model):
    self.model = model

  @tf.function(
      input_signature=(
        tf.TensorSpec(shape=[None,None,3], dtype=tf.float32),
        tf.TensorSpec(shape=[], dtype=tf.int32),
        tf.TensorSpec(shape=[], dtype=tf.float32),)
  )
  def __call__(self, img, steps, step_size):
      print("Tracing")
      loss = tf.constant(0.0)
      for n in tf.range(steps):
        with tf.GradientTape() as tape:
          # Aca se necesita el gradiente relativo a la imagen (`img`)
          # La variable `GradientTape` solo puede ver `tf.Variable`s por defecto
          tape.watch(img)
          loss = calc_loss(img, self.model)

        # Calculando la función de perdida con los gradientes relacionados a los pixeles de la imagen de entrada.
        gradients = tape.gradient(loss, img)

        # Normalizando los gradientes.
        gradients /= tf.math.reduce_std(gradients) + 1e-8 

        # En gradiente ascendente, la variable "loss" es para maximizar el incremento sobre las neuronas de las capas involucradas.
        # Es posible incrementar manualmente el grado de los gradientes en la imagen, esto porque tienen el mismo tamaño
        img = img + gradients*step_size
        img = tf.clip_by_value(img, -1, 1)

      return loss, img
```

```python
# Esta línea permite 'guardar' el modelo generado en una variable para luego ser utilizada
deepdream = DeepDream(dream_model)
```

### 4. Tratamiento de la imagen sobre la red neuronal

```python
# Bucle principal
def run_deep_dream_simple(img, steps=100, step_size=0.01):
  # Convertir desde el formato uint8 al rango esperado por el modelo.
  img = tf.keras.applications.inception_v3.preprocess_input(img)
  img = tf.convert_to_tensor(img)
  step_size = tf.convert_to_tensor(step_size)
  steps_remaining = steps
  step = 0
  while steps_remaining:
    if steps_remaining>100:
      run_steps = tf.constant(100)
    else:
      run_steps = tf.constant(steps_remaining)
    steps_remaining -= run_steps
    step += run_steps

    loss, img = deepdream(img, run_steps, tf.constant(step_size))

    display.clear_output(wait=True)
    show(deprocess(img))
    print ("Step {}, loss {}".format(step, loss))

  result = deprocess(img)
  display.clear_output(wait=True)
  show(result)

  return result
```

```python
# Y luego, es posible generar un primer vistazo al desarrollo de esta red sobre la imagen cargada
dream_img = run_deep_dream_simple(img=original_img, steps=100, step_size=0.01)
```

#### Con todo lo anterior, el modelo ya ha sido capaz de tratar una imagen desde internet y pasar por la red neuronal¡¡¡¡

### 5. Realizando mejoras al modelo

Si bien, para un primer intento ha resultado bastante bien, hay algunos problemas que se pueden generar con esta predicción:

1. La salida es ruidosa (esto podría solucionarse con una función de pérdida de la forma [tf.image.total_variation](https://www.tensorflow.org/api_docs/python/tf/image/total_variation) lo que permitirá dar una mejor resolución sobre el tratamiento de los pixeles de la imagen).
2. La imagen puede ser de baja resolución, por lo tanto, al pasar por el modelo se entrenará más rápido pero se perderán detalles de la misma.
3. Los patrones parecen estar ocurriendo con la misma granularidad.

Un enfoque que aborda todos estos problemas es aplicar un ascenso de gradiente a diferentes escalas. Esto permitirá que los patrones generados a escalas más pequeñas se incorporen a patrones a escalas más altas y se completen con detalles adicionales.

Para hacer esto, es posible realizar el enfoque de ascenso de gradiente anterior, luego aumentar el tamaño de la imagen (que se conoce como una octava) y repetir este proceso para varias octavas.

```python
# Subiendo la octava
import time
start = time.time()

OCTAVE_SCALE = 1.40

img = tf.constant(np.array(original_img))
base_shape = tf.shape(img)[:-1]
float_base_shape = tf.cast(base_shape, tf.float32)

for n in range(-5, 6):
  new_shape = tf.cast(float_base_shape*(OCTAVE_SCALE**n), tf.int32)

  img = tf.image.resize(img, new_shape).numpy()

  img = run_deep_dream_simple(img=img, steps=50, step_size=0.01)

display.clear_output(wait=True)
img = tf.image.resize(img, base_shape)
img = tf.image.convert_image_dtype(img/255.0, dtype=tf.uint8)
show(img)

end = time.time()
end-start
```

#### Con lo cual, es posible apreciar una primera mejora a los bordes y detalles de este nuevo paso.

### Adicional: Mejora con el uso de mosaicos

Una cosa a considerar es que a medida que la imagen aumenta de tamaño, también lo hará el tiempo y la memoria necesarios para realizar el cálculo del gradiente. La implementación de octava anterior no funcionará en imágenes muy grandes o con muchas octavas.

Para evitar este problema, se puede dividir la imagen en mosaicos y calcular el degradado de cada mosaico.

La aplicación de cambios aleatorios a la imagen antes de cada cálculo de mosaico evita que aparezcan costuras de mosaico.

```python
# Adicional, uso de mosaicos
def random_roll(img, maxroll):
  # Este paso se realiza de manera 'aleatoria' sobre la imagen para evitar los límites del mosaico.
  shift = tf.random.uniform(shape=[2], minval=-maxroll, maxval=maxroll, dtype=tf.int32)
  img_rolled = tf.roll(img, shift=shift, axis=[0,1])
  return shift, img_rolled
```

```python
shift, img_rolled = random_roll(np.array(original_img), 512)
show(img_rolled)
```

Desde donde se puede apreciar la aparición de un "filtro" en forma de mosaico dentro de la imagen inicial

```python
class TiledGradients(tf.Module):
  def __init__(self, model):
    self.model = model

  @tf.function(
      input_signature=(
        tf.TensorSpec(shape=[None,None,3], dtype=tf.float32),
        tf.TensorSpec(shape=[2], dtype=tf.int32),
        tf.TensorSpec(shape=[], dtype=tf.int32),)
  )
  def __call__(self, img, img_size, tile_size=512):
    shift, img_rolled = random_roll(img, tile_size)

    # Inicializa la variable que contiene los gradientes en ceros.
    gradients = tf.zeros_like(img_rolled)

    # Estas líneas permiten omitir el ultimo mosaico, siempre y cuando no quede solo uno (y por tanto no pueda obtener uno).
    xs = tf.range(0, img_size[1], tile_size)[:-1]
    if not tf.cast(len(xs), bool):
      xs = tf.constant([0])
    ys = tf.range(0, img_size[0], tile_size)[:-1]
    if not tf.cast(len(ys), bool):
      ys = tf.constant([0])

    for x in xs:
      for y in ys:
        # Calcula los gradientes para cada uno de los mosaicos.
        with tf.GradientTape() as tape:
          # Acá se necesita el gradiente relativo a la imagen (`img`)
          # La variable `GradientTape` solo puede ver `tf.Variable`s por defecto
          tape.watch(img_rolled)

          # Extrae un mosaico de la imagen.
          img_tile = img_rolled[y:y+tile_size, x:x+tile_size]
          loss = calc_loss(img_tile, self.model)

        # Actualiza (incorpora) los gradientes sobre la imagen inicial.
        gradients = gradients + tape.gradient(loss, img_rolled)

    # Deshace los cambios aplicados sobre la imagen y sus gradientes
    gradients = tf.roll(gradients, shift=-shift, axis=[0,1])

    # Normaliza los gradientes.
    gradients /= tf.math.reduce_std(gradients) + 1e-8 

    return gradients
```

```python
get_tiled_gradients = TiledGradients(dream_model)
```

Al combinar estos pasos se genera una implementación de sueño profundo escalable y con reconocimiento de octava

```python
def run_deep_dream_with_octaves(img, steps_per_octave=100, step_size=0.01, 
                                octaves=range(-2,3), octave_scale=1.3):
  base_shape = tf.shape(img)
  img = tf.keras.utils.img_to_array(img)
  img = tf.keras.applications.inception_v3.preprocess_input(img)

  initial_shape = img.shape[:-1]
  img = tf.image.resize(img, initial_shape)
  for octave in octaves:
    # Escala la imagen por medio de la técnica de la octava
    new_size = tf.cast(tf.convert_to_tensor(base_shape[:-1]), tf.float32)*(octave_scale**octave)
    new_size = tf.cast(new_size, tf.int32)
    img = tf.image.resize(img, new_size)

    for step in range(steps_per_octave):
      gradients = get_tiled_gradients(img, new_size)
      img = img + gradients*step_size
      img = tf.clip_by_value(img, -1, 1)

      if step % 10 == 0:
        display.clear_output(wait=True)
        show(deprocess(img))
        print ("Octave {}, Step {}".format(octave, step))

  result = deprocess(img)
  return result
```

```python
img = run_deep_dream_with_octaves(img=original_img, step_size=0.01)

display.clear_output(wait=True)
img = tf.image.resize(img, base_shape)
img = tf.image.convert_image_dtype(img/255.0, dtype=tf.uint8)
show(img)
```

#### Con este nuevo tratamiento se obtiene un resultado mucho mejor.
